## Data manipulation

### Subtask:
Perform some basic data manipulation tasks using pandas and numpy, such as filtering or creating new columns.


Create a new column for yearly income, filter the DataFrame based on this new column, and calculate the average years at the company for the filtered data.



```python
df['Income_per_Year'] = df['MonthlyIncome'] * 12
```

```python
df_high_income = df[df['Income_per_Year'] > 70000].copy()
```

```python
avg_years_high_income = df_high_income['YearsAtCompany'].mean()
```

## Visualize manipulated data

### Subtask:
Create visualizations of the manipulated data.


Create the requested visualizations using matplotlib and seaborn to understand the data distributions and relationships between variables.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Histogram of Income_per_Year from df
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Income_per_Year', kde=True)
plt.title('Distribution of Yearly Income')
plt.xlabel('Yearly Income')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Box plot comparing Income_per_Year for employees who stayed versus those who left
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Attrition', y='Income_per_Year', palette='coolwarm')
plt.title('Yearly Income vs. Attrition')
plt.xlabel('Attrition')
plt.ylabel('Yearly Income')
plt.show()
```

```python
# 3. Scatter plot showing the relationship between TotalWorkingYears and Income_per_Year for df_high_income
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_high_income, x='TotalWorkingYears', y='Income_per_Year', alpha=0.6, color='purple')
plt.title('Total Working Years vs. Yearly Income (High Income Earners)')
plt.xlabel('Total Working Years')
plt.ylabel('Yearly Income')
plt.show()
```
