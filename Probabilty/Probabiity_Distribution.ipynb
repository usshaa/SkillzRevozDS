````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import binom, poisson, norm, expon, lognorm
````

````python
sns.set(style="whitegrid")
````

````python
# Plot helper
def show_distribution(x, y, title, xlabel, ylabel="Probability"):
    plt.figure(figsize=(8, 4))
    plt.plot(x, y, marker='o')
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.show()
````

## 1️⃣ **Binomial Distribution**

> "How many people buy cupcakes out of 10?"

````python
# Parameters: 10 people, 0.3 chance each buys a cupcake
n = 10
p = 0.3
x = np.arange(0, n+1)
y = binom.pmf(x, n, p)  # PMF = Probability Mass Function
````

````python
show_distribution(x, y, "Binomial Distribution: Cupcake Buyers", "Number of Buyers")
````

## 2️⃣ **Poisson Distribution**

> "How many customers enter a shop in an hour?"

````python
# Average 5 customers per hour
lambda_ = 5
x = np.arange(0, 15)
y = poisson.pmf(x, mu=lambda_)
````

````python
show_distribution(x, y, "Poisson Distribution: Customers in an Hour", "Customers")
````

## 3️⃣ **Normal Distribution**

> "Cupcake weight around 100g with little variation"

````python
mean = 100
std_dev = 10
x = np.linspace(60, 140, 100)
y = norm.pdf(x, mean, std_dev)  # PDF = Probability Density Function
````

````python
show_distribution(x, y, "Normal Distribution: Cupcake Weights", "Weight (grams)")
````

## 4️⃣ **Exponential Distribution**

> "Time until the next customer comes in"

````python
scale = 5  # Average wait time = 5 minutes
x = np.linspace(0, 20, 100)
y = expon.pdf(x, scale=scale)
````

````python
show_distribution(x, y, "Exponential Distribution: Wait Time for Next Customer", "Time (minutes)")
````

## 5️⃣ **Log-Normal Distribution**

> "Skewed product prices in a store"

````python
mu = 2.5
sigma = 0.6
x = np.linspace(0.01, 20, 100)
y = lognorm.pdf(x, s=sigma, scale=np.exp(mu))
````

````python
show_distribution(x, y, "Log-Normal Distribution: Product Prices", "Price (₹)")
````

### 📌 Summary of When to Use Each Distribution


```
📦 BINOMIAL: Used for YES/NO counts (e.g., survey responses)
📦 POISSON: Used for counting events over time/space (e.g., visitors per hour)
📦 NORMAL: Used when values cluster around a mean (e.g., heights, scores)
📦 EXPONENTIAL: Used for time until next event (e.g., wait time)
📦 LOG-NORMAL: Used for skewed, always-positive values (e.g., prices, incomes)
```
