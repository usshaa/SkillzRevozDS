## Why Probability for Data Analytics?

* Data is uncertain → probability handles uncertainty
* Predictive modeling relies on probability
* Risk analysis, customer behavior, product success rate = all require probability

```python
import random

# Coin flip simulation
outcomes = [random.choice(['Heads', 'Tails']) for _ in range(100)]
heads_prob = outcomes.count('Heads') / len(outcomes)
print(f"Probability of Heads ≈ {heads_prob}")
```

### Basic Probability Concepts

#### 🧠 Concepts:

* Sample space, event, probability of an event
* Union, intersection, conditional probability
* Independent vs dependent events

#### 🧪 Code: Basic Set Probabilities

```python
A = set([1, 2, 3, 4])
B = set([3, 4, 5, 6])

print("P(A or B):", len(A.union(B)) / 6)
print("P(A and B):", len(A.intersection(B)) / 6)
```

### Probability Distributions

* Discrete: Binomial, Poisson
* Continuous: Normal, Exponential

#### 🔢 Visual Comparison

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram of normal vs binomial
sns.histplot(np.random.normal(0, 1, 1000), kde=True, label='Normal')
sns.histplot(np.random.binomial(10, 0.5, 1000), kde=True, color='red', label='Binomial')
plt.legend()
plt.title("Normal vs Binomial Distribution")
plt.show()
```

### Binomial Distribution

#### 🎓 Theory:

* n = number of trials
* p = probability of success
* Discrete values → count of successes

#### ✅ Use Case: "What’s the chance of guessing 6 correct answers out of 10?"

#### 🧪 Python Example:

```python
from scipy.stats import binom
import matplotlib.pyplot as plt

n, p = 10, 0.5
x = range(0, 11)
pmf = binom.pmf(x, n, p)

plt.bar(x, pmf)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.show()
```

### ⏰ Poisson Distribution

#### 🎓 Theory:

* Models rare events (e.g., number of calls per minute)
* λ (lambda) = average rate

#### ✅ Use Case: "Expected number of website crashes per week?"

#### 🧪 Python Code:

```python
from scipy.stats import poisson

mu = 3  # avg events per interval
x = range(0, 10)
pmf = poisson.pmf(x, mu)

plt.bar(x, pmf)
plt.title("Poisson Distribution (λ = 3)")
plt.xlabel("Number of Events")
plt.ylabel("Probability")
plt.show()
```

### Central Limit Theorem (CLT)

* Sampling distribution of sample means → tends toward normal distribution, no matter the original distribution

#### ✅ Averages of customer wait times

#### 🧪 Python Code:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(0)

# Exponential distribution (non-normal)
data = np.random.exponential(scale=2, size=1000)

# Sampling means
sample_means = [np.mean(np.random.choice(data, size=50)) for _ in range(1000)]

sns.histplot(sample_means, kde=True)
plt.title("Central Limit Theorem: Distribution of Sample Means")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```
